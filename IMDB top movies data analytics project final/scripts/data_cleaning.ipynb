```python
import pandas as pd
```

```python
df = pd.read_csv('IMDB top 1000.csv')
```

```python
df
```

```python
df2 = df.drop_duplicates()
df2
```

```python
df2["Title"] = df2["Title"].str.split().map(lambda x: x[1:]).str.join(" ")
```

```python
df2
```

```python
df2.rename(columns={"Unnamed: 0": "Rank"}, inplace=True)
```

```python
df2["Rank"] = df2["Rank"] + 1
```

```python
df2
```

```python
df2["Duration"] = df2["Duration"].str.split().map(lambda x: x[0])
```

```python
df2.rename(columns={"Duration": "Duration(minutes)"}, inplace=True)
```

```python
df2
```

```python
certification_mapping = {
    "G": "G",
    "GP": "PG",
    "PG": "PG",
    "PG-13": "PG-13",
    "TV-PG": "PG-13",
    "TV-14": "PG-13",
    "R": "R",
    "TV-MA": "R",
    "Approved": "Unrated",
    "Passed": "Unrated",
    "Not Rated": "Unrated",
    "": "Unrated"  # Handle blank entries
}

df2['Certificate'] = df['Certificate'].replace(certification_mapping)
```

```python
df2
```

```python
df2["Certificate"].fillna("Unrated", inplace=True)
```

```python
df2
```

```python
df2["Year"] = df2["Title"].str.split().map(lambda x: int(x[-1][1:-1]))
```

```python
df2
```

```python
df2["Title"] = df2["Title"].str.split().map(lambda x: x[:-1]).str.join(" ")
```

```python
df2
```

```python
df2[["Genre1", "Genre2", "Genre3"]] = df2["Genre"].str.split(',',2, expand=True)
```

```python
df2
```

```python
df2.drop(columns=["Rank"], inplace=True)
```

```python
df2["Votes"] = df2["Info"].str.split().map(lambda x: int(str(x[1]).replace(',', '')))
```

```python
df2
```

```python
df2["Gross(in million dollars)"] = df2["Info"].str.split().map(lambda x: float(x[4].replace('$', "").replace("M", "")) if len(x) > 3 else None)
```

```python
df2
```

```python
df2["Metascore"] = df["Metascore"].replace("NaN", None)
```

```python
df2
```

```python
# df2["Cast"].str.split('|')[0][0].replace("Director: ", "")
directors = []
for i in df2["Cast"]:
    inter = i.split('|')[0]
    if inter.startswith("Director: "):
        interarr = inter.replace("Director: ", "").strip().split(",")
        directors.append([interarr[0], None, None])
    else:
        interarr = inter.replace("Directors: ", "").split(",")
        newarr = []
        for i in interarr:
            newarr.append(i.strip())
        if len(newarr) == 2:
            newarr.append(None)
        directors.append(newarr) 

print(directors)

```

```python
df2[["Director1", "Director2", "Director3"]] = directors
```

```python
df2
```

```python
# df2["Cast"].str.split('|')[0][0].replace("Director: ", "")
stars = []
for i in df2["Cast"]:
    inter = i.split('|')[1]
    interarr = inter.replace("Stars: ", "").split(",")
    newarr = []
    for i in interarr:
        newarr.append(i.strip())

    stars.append(newarr)
```

```python
df2[["Actor1", "Actor2", "Actor3", "Actor4"]] = stars
```

```python
df2
```

```python
df2
```

```python
df2
```

```python
df3 = df2.drop_duplicates()
df3
```

```python
df3.to_csv("IMDB top 1000 clean new.csv", index=False)
```
